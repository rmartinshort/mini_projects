### This notebook uses a couple of different methods to classify the sklearn wine data  

It is mainly intended as an exploration of the functionality of tensor flow, but it contains exploration of a variety of other classification algoithms in an attempt to see which works best. This was also used to learn the basic functionalty of Keras, and to compare it with tensorflow's Layers package

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
```

```python
print(wine.DESCR)
```

### Lets first visualize the data

```python
colnames = wine.feature_names
wine_df = pd.DataFrame(wine.data,columns=colnames)
target_df = pd.DataFrame(wine.target,columns=['target'])
```

```python
wine_df.head()
```

```python
full_df = pd.concat([wine_df,target_df],axis=1)
full_df.head()
sns.pairplot(full_df,hue='target')
```

This plot shows that the classes are fairly well separated - we might do a decent job classifying with a logistic regression infact. At any rate, it would be interesting to see the extent to which improvements in classification result from increasing complexity in the ML algorithm choice 

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
```

### Classification with tensorflow 

Here we'll use a basic, fully connected ANN. Tensorflow allows much more complex constructions, but this is probably fine for now

```python
from tensorflow.contrib.layers import fully_connected
import tensorflow as tf
```

```python
#Do a train-test split
X_train, X_test, y_train, y_test = train_test_split(wine.data,wine.target,test_size=0.3)
```

```python
scaler = StandardScaler()
scaled_x_train = scaler.fit_transform(X_train)
scaled_x_test = scaler.transform(X_test)
np.shape(scaled_x_train)
```

```python
#Do one-hot encoding
onehot_y_train = pd.get_dummies(y_train).as_matrix() 
onehot_t_test = pd.get_dummies(y_test).as_matrix()
```

```python
#Set up DNN parameters
num_feat = len(colnames)
#Lets have three hidden layers. Maybe too many for this small dataset?
num_hidden1 = 20
num_hidden2 = 20
num_hidden3 = 20
#Number of outputs
num_outputs = len(target_df['target'].unique())
print(num_outputs)
#This learning rate will need to be adjusted
learning_rate = 0.01
#Training steps in model
training_steps = 5000
```

```python
#Set up variables
X = tf.placeholder(tf.float32,shape=[None,num_feat])
y_true = tf.placeholder(tf.float32,shape=[None,num_outputs])

act_function = 'tanh'

#Can choose various activation functions
#In this case the choice doen't need to make much difference
if act_function == 'relu':
    actf = tf.nn.relu #relu activation function
elif act_function == 'tanh':
    actf = tf.nn.tanh
elif act_function == 'sigmoid':
    actf = tf.nn.sigmoid
```

```python
#Set up layers - this is where the layers package is useful because it allows constuction
#of the network geometry
hidden1 = fully_connected(X,num_hidden1,activation_fn=actf)
hidden2 = fully_connected(hidden1,num_hidden2,activation_fn=actf)
hidden3 = fully_connected(hidden2,num_hidden3,activation_fn=actf)
output = fully_connected(hidden3,num_outputs)
```

```python
#Define a loss function - lots of choice here
#loss = tf.losses.softmax_cross_entropy(onehot_labels=y_true,logits=output)
loss = tf.losses.mean_squared_error(labels=y_true,predictions=output)
```

```python
#Adam optimizer seems to be a good choice for testing
optimizer = tf.train.AdamOptimizer(learning_rate)
#optimizer = tf.train.AdadeltaOptimizer(learning_rate) This one didn't work so well
```

```python
train = optimizer.minimize(loss)
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    
    sess.run(init)
    for i in range(training_steps):
        
        sess.run(train,feed_dict={X:scaled_x_train,y_true:onehot_y_train})
    
    logits = output.eval(feed_dict={X:scaled_x_test})
    
    pred = tf.argmax(logits,axis=1)
    
    results = pred.eval()
```

```python
print(classification_report(results,y_test))
```

```python
print(confusion_matrix(results,y_test))
```
